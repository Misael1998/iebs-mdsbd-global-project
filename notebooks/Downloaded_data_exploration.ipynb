```python
import pandas as pd
```

```python
nrr_features = pd.read_csv('../data/downloaded/nrr/nrr_features.csv')
```

```python
sentinel_features = pd.read_csv('../data/downloaded/sentinel/features.csv')
```

```python
metadata = pd.read_csv('../data/metadata.csv')
```

```python
train_labels = pd.read_csv('../data/train_labels.csv')
```

```python
na = nrr_features.columns[nrr_features.isna().any()].tolist()
```

```python
nrr_features = nrr_features.drop(na, axis=1)
```

```python
nrr_features = nrr_features.drop(['Unnamed: 0'], axis=1)
```

```python
sentinel_features = sentinel_features.drop(['Unnamed: 0'], axis=1)
```

```python
nrr_features.head().transpose()
```

```python
merged = pd.merge(sentinel_features, nrr_features, on='uid', how='inner')
```

```python
len(merged)
```

```python
merged = pd.merge(merged, metadata, on='uid', how='inner')
```

```python
len(merged)
```

```python
merged['split'].value_counts()
```

```python
merged = pd.merge(merged, train_labels, on='uid', how='inner')
```
